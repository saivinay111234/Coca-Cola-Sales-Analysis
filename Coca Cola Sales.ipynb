```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Coca Cola Sales Data.csv")
print(data)
```

```python
# Data Preprocessing
data.isnull().sum()
```

```python
data = data.rename(columns={
    'Days to \nDeliver': 'Days to Deliver',
    'Delivery \nCompany': 'Delivery Company',
    'Price per \nUnit' : 'Price per Unit'
})

data.columns
```

```python
data.dtypes
```

```python
data.describe()

# Analysis Findings:
# The Average no of days to deliver the product is 37 days. while the lowest no of days is 15 days and Max time to deliver the product is 60 days.
```

```python
df = data
```

```python
# Remove non-numeric characters and convert to numeric
df['Price per Unit'] = df['Price per Unit'].str.replace('[\$,]', '', regex=True).astype(float)
df['Units Sold'] = df['Units Sold'].str.replace(',', '').astype(int)
df['Total Sales'] = df['Total Sales'].str.replace('[\$,]', '', regex=True).astype(float)
df['Operating Profit'] = df['Operating Profit'].str.replace('[\$,]', '', regex=True).astype(float)
df['Operating Margin'] = df['Operating Margin'].str.replace('%', '', regex=True).astype(float) / 100

# Display the updated DataFrame
df
```

```python
df.dtypes
```

```python
cat_col = ['Retailer', 'Region', 'State', 'City', 'Beverage Brand',	'Delivery Company']
for i in cat_col:
    print(f'Unique values in the "{i}" column are : {df[i].unique()}')
    print(df[i].value_counts())
    print("\n")
```

```python
df1 = df
```

```python
# Feature Engineering
df1['Sales per Unit'] = df1['Total Sales'] / df1['Units Sold']
df1['Profit per Unit'] = df1['Operating Profit'] / df1['Units Sold']
df1['Delivery Efficiency'] = df1['Total Sales'] / df1['Days to Deliver']
df1['Invoice Date'] = pd.to_datetime(df1['Invoice Date'])
df1['Year'] = df1['Invoice Date'].dt.year
df1['Month'] = df1['Invoice Date'].dt.month
df1['Profit Margin'] = (df1['Operating Profit'] / df1['Total Sales']) * 100

df1
```

```python
# Sales Over Time: Analyze sales data over time (using the Invoice Date) to identify trends, seasonal patterns, and peak sales periods.

df1['Invoice Date'] = pd.to_datetime(df1['Invoice Date'])

# Group the data by 'Invoice Date' and calculate total sales
total_sales_over_time = df1.groupby('Invoice Date')['Total Sales'].sum().reset_index()

# Plot the total sales over time
plt.figure(figsize=(18, 8))
sns.lineplot(data=total_sales_over_time, x='Invoice Date', y='Total Sales', marker='o', color='green')
plt.title('Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# we cant clearly see the trend here, except there been a increase in sales from last day of the month to first day of the next month i.e. 31 January 
# to 1st Feb. To see the trend we can use 3 methods:

# Rolling Average Plot: This smooths out short-term fluctuations and highlights longer-term trends.
# Seasonal Decomposition Plot: This breaks down the time series into trend, seasonal, and residual components.
# Monthly/Quarterly Aggregated Plot: Aggregating the data by month or quarter can make seasonal patterns more apparent.
```

```python
%pip install statsmodels 
```

```python
# Rolling average plot
# Add a column for rolling average of total sales (e.g., 30-day rolling average)
total_sales_over_time['Rolling Average'] = total_sales_over_time['Total Sales'].rolling(window=30).mean()

# Plot the total sales and rolling average over time
plt.figure(figsize=(18, 8))
sns.lineplot(data=total_sales_over_time, x='Invoice Date', y='Total Sales', marker='o', color='blue', label='Daily Sales')
sns.lineplot(data=total_sales_over_time, x='Invoice Date', y='Rolling Average', marker='', color='red', label='30-Day Rolling Average')
plt.title('Total Sales Over Time with Rolling Average')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()


# 2.Seasonal Decomposition Plot
# Seasonal decomposition using moving averages can break down the time series into its trend, seasonal, and residual components. 

from statsmodels.tsa.seasonal import seasonal_decompose

# Decompose the time series
decomposition = seasonal_decompose(total_sales_over_time.set_index('Invoice Date')['Total Sales'], model='additive', period=30)

# Plot the decomposition
plt.figure(figsize=(14, 10))
decomposition.plot()
plt.tight_layout()
plt.show()


# Sales by Month-Quaterly
# Create a column for year-month
total_sales_over_time['YearMonth'] = total_sales_over_time['Invoice Date'].dt.to_period('M')

# Group the data by year-month and calculate total sales
monthly_sales = total_sales_over_time.groupby('YearMonth')['Total Sales'].sum().reset_index()
monthly_sales['YearMonth'] = monthly_sales['YearMonth'].dt.to_timestamp()

# Plot the monthly aggregated sales
plt.figure(figsize=(18, 8))
sns.lineplot(data=monthly_sales, x='YearMonth', y='Total Sales', marker='o', color='green')
plt.title('Total Sales by Month')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Quaterly Sales Plot
df1['Invoice Date'] = pd.to_datetime(df1['Invoice Date'])

# Add a column for the quarter
df1['Quarter'] = df1['Invoice Date'].dt.to_period('Q')

# Group the data by year-quarter and calculate total sales
quarterly_sales = df1.groupby('Quarter')['Total Sales'].sum().reset_index()
quarterly_sales['Quarter'] = quarterly_sales['Quarter'].dt.to_timestamp()

# Plot the quarterly aggregated sales
plt.figure(figsize=(18, 8))
sns.lineplot(data=quarterly_sales, x='Quarter', y='Total Sales', marker='o', color='purple')
plt.title('Total Sales by Quarter')
plt.xlabel('Quarter')
plt.ylabel('Total Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 6. Customer Behavior and Preferences
# Retailer Purchase Patterns: Study the purchase patterns of different retailers to understand their preferences and buying behavior.

# To analyze customer behavior and preferences through retailer purchase patterns, we can focus on several aspects:
# Potential Analyses:

# 1. Purchase Frequency by Retailer: Determine how often each retailer makes purchases.
# Group by Retailer and count the number of purchases
purchase_frequency = df['Retailer'].value_counts()

# Plotting the purchase frequency by retailer
plt.figure(figsize=(18, 8))
sns.barplot(x=purchase_frequency.index, y=purchase_frequency.values, hue = purchase_frequency.index, legend = 0, palette='viridis')
plt.title('Purchase Frequency by Retailer')
plt.xlabel('Retailer')
plt.ylabel('Number of Purchases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# 2. Popular Products by Retailer: Identify which products are most commonly purchased by each retailer.
# Group by Retailer and Beverage Brand, then count the number of purchases
popular_products = df.groupby(['Retailer', 'Beverage Brand']).size().unstack(fill_value=0)

# Plotting the heatmap
plt.figure(figsize=(18, 8))
sns.heatmap(popular_products, cmap='viridis', annot=True, fmt='d')
plt.title('Popular Products by Retailer')
plt.xlabel('Beverage Brand')
plt.ylabel('Retailer')
plt.tight_layout()
plt.show()

# 3. Average Order Value by Retailer: Calculate the average order value for each retailer to understand their spending behavior.
# Calculate average order value by Retailer
average_order_value = df.groupby('Retailer')['Total Sales'].mean()

# Plotting the average order value by retailer
plt.figure(figsize=(18, 8))
sns.barplot(x=average_order_value.index, y=average_order_value.values, hue = average_order_value.index, palette='magma', legend = 0)
plt.title('Average Order Value by Retailer')
plt.xlabel('Retailer')
plt.ylabel('Average Order Value')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 7. Delivery and Logistics Insights
# Correlation Between Days to Deliver and Sales: Investigate if longer delivery times impact the number of units sold or total sales.

# To investigate the correlation between days to deliver and sales, we can perform below steps:
# 1. Correlation Analysis: Calculate the correlation coefficient between "Days to Deliver" and "Total Sales" or "Units Sold".
# 2. Scatter Plot: Visualize the relationship between "Days to Deliver" and "Total Sales" or "Units Sold" using scatter plots.
# 3. Regression Analysis: Fit a regression line to understand the nature and strength of the relationship.

# Calculate the correlation coefficient
correlation_sales = df1['Days to Deliver'].corr(df1['Total Sales'])
correlation_units = df1['Days to Deliver'].corr(df1['Units Sold'])

print(f"Correlation between Days to Deliver and Total Sales: {correlation_sales}")
print(f"Correlation between Days to Deliver and Units Sold: {correlation_units}")


# Scatter plot with regression line for Days to Deliver vs Total Sales
plt.figure(figsize=(18, 8))
sns.regplot(data=df1, x='Days to Deliver', y='Total Sales', line_kws={"color":"r","alpha":0.7,"lw":2})
plt.title('Days to Deliver vs Total Sales with Regression Line')
plt.xlabel('Days to Deliver')
plt.ylabel('Total Sales')
plt.show()

# Scatter plot with regression line for Days to Deliver vs Units Sold
plt.figure(figsize=(18, 8))
sns.regplot(data=df1, x='Days to Deliver', y='Units Sold', line_kws={"color":"r","alpha":0.7,"lw":2})
plt.title('Days to Deliver vs Units Sold with Regression Line')
plt.xlabel('Days to Deliver')
plt.ylabel('Units Sold')
plt.show()

```

```python
def adfuller_test(sales):
    result=adfuller(sales)
    labels = ['ADF Test Statistic','p-value','#Lags Used','Number of Observations']
    for value,label in zip(result,labels):
        print(label+' : '+str(value) )

    if result[1] <= 0.05:
        print("strong evidence against the null hypothesis(Ho), reject the null hypothesis. Data is stationary")
    else:
        print("weak evidence against null hypothesis,indicating it is non-stationary ")

adfuller_test(df4['Total Sales'])
```
